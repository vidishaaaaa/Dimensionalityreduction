```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import seaborn as sns
import plotly.express as px
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import MDS
```

```python
X,y=load_digits(return_X_y=True)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(X[1].reshape(8,8))
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=3,sharex=True,sharey=True, figsize=(10,10))
axes[0,0].imshow(X[1].reshape(8,8))
axes[0,1].imshow(X[0].reshape(8,8))
axes[0,2].imshow(X[3].reshape(8,8))
axes[1,2].imshow(X[2].reshape(8,8))
axes[1,1].imshow(X[5].reshape(8,8))
axes[1,0].imshow(X[4].reshape(8,8))
plt.show()
plt.tight_layout()
```

Local Linear Embedding

```python
lle= LocallyLinearEmbedding(n_components=2,n_neighbors=15)
X_lle=lle.fit_transform(X,y=y) # returns two new column

data= pd.DataFrame({'LLE1':X_lle[:,0],'LLE2':X_lle[:,1],'class':y})
plt.figure(figsize=(10,10))
sns.scatterplot(x='LLE1',y='LLE2',hue='class',data=data,palette='tab10')
plt.show()
```

3d

```python
lle= LocallyLinearEmbedding(n_components=3,n_neighbors=15)
X_lle=lle.fit_transform(X,y=y) # returns three new column
data=pd.DataFrame({'LLE1':X_lle[:,0],'LLE2':X_lle[:,1],'LLE3':X_lle[:,2],'class':y})
fig=px.scatter_3d(data,x='LLE1',y='LLE2',z='LLE3',color='class')
fig.show()
```

Principal Component analysis


```python
mypca=PCA(n_components=2)
X_pca=mypca.fit_transform(X)
data=pd.DataFrame({'PCA1':X_pca[:,0],'PCA2':X_pca[:,1],'class':y})
sns.scatterplot(x='PCA1',y='PCA2',hue='class',data=data,palette='tab10')
plt.show()

```

3d


```python
myPCA=PCA().fit(X)
X_pca=myPCA.transform(X)
data= pd.DataFrame({'PCA1':X_pca[:,0],'PCA2':X_pca[:,1],'PCA3':X_pca[:,2],'class':y})
fig=px.scatter_3d(data,x='PCA1',y='PCA2',z='PCA3',color='class')
fig.show()
```

fisher discriminant analysis

```python
lda= LinearDiscriminantAnalysis(n_components=2)
x_lda=lda.fit_transform(X,y=y)
data=pd.DataFrame({'LDA1':x_lda[:,0],'LDA2':x_lda[:,1],'class':y})
fig=plt.figure(figsize=(10,10))
sns.scatterplot(x='LDA1',y='LDA2',hue='class',data=data,palette='tab10')
```

3d

```python
lda= LinearDiscriminantAnalysis(n_components=3)
x_lda=lda.fit_transform(X,y=y)
data=pd.DataFrame({'LDA1':x_lda[:,0],'LDA2':x_lda[:,1],'LDA3':x_lda[:,2],'class':y})
fig=px.scatter_3d(data,x='LDA1',y='LDA2',z='LDA3',color='class')
fig.show()
```
